```python
import gensim
```

```python
DIR = 'GoogleNews-vectors-negative300.bin'
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format(DIR, binary=True)
```

```python
model.init_sims(replace=True)
```

```python
model.wv['cat']
```

```python
model.most_similar(negative=['run', 'walk'], topn=1)
```

```python
model.doesnt_match("student professor homework lecturer".split())
```

```python
model.similarity('apple', 'computer')
```
